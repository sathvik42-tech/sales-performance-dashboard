```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/processed/cleaned_sales.csv")
```

```python
df.head()
```

```python
total_sales = df["Sales"].sum()
print(f"Total Sales: ${total_sales:,.2f}")
```

```python
total_profit = df["Profit"].sum()
print(f"Total Profit: ${total_profit:,.2f}")
```

```python
average_sales = df["Sales"].mean()
print(f"Average Sales: ${average_sales:.2f}")
```

```python
total_orders = len(df)
print("Total Orders:", total_orders)
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

print(category_sales)
```

```python
total_quantity = df["Quantity"].sum()

print("Total Quantity Sold:", total_quantity)
```

```python
category_sales = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)

category_sales
```

```python
plt.figure(figsize=(8,5))

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.xticks(rotation=0)
plt.savefig("../images/charts/sales_by_category.png", dpi=300)

plt.show()
```

```python
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)

category_profit
```

```python
plt.figure(figsize=(8,5))

category_profit.plot(kind="bar")

plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")

plt.xticks(rotation=0)

plt.savefig("../images/charts/profit_by_region.png", dpi=300)
plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)

region_sales
```

```python
plt.figure(figsize=(8,5))

region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.xticks(rotation=0)

plt.show()
```

```python
region_profit = df.groupby("Region")["Profit"].sum().sort_values(ascending=False)

region_profit
```

```python
plt.figure(figsize=(8,5))

region_profit.plot(kind="bar")

plt.title("Profit by Region")
plt.xlabel("Region")
plt.ylabel("Profit")

plt.xticks(rotation=0)

plt.savefig("../images/charts/profit_by_region.png", dpi=300)

plt.show()
```

```python
print("Total Sales:", df["Sales"].sum())
print("Total Profit:", df["Profit"].sum())

print("\nTop Category:")
print(df.groupby("Category")["Sales"].sum().sort_values(ascending=False))

print("\nTop Region:")
print(df.groupby("Region")["Profit"].sum().sort_values(ascending=False))

print("\nTop Segment:")
print(df.groupby("Segment")["Sales"].sum().sort_values(ascending=False))
```

```python
state_sales = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

state_sales
```

```python
plt.figure(figsize=(10,6))

state_sales.plot(kind="bar")

plt.title("Top 10 States by Sales")
plt.xlabel("State")
plt.ylabel("Sales")

plt.xticks(rotation=45)
plt.savefig("../images/charts/top10_states.png", dpi=300)

plt.show()
```

```python
plt.figure(figsize=(7,7))

segment_sales = df.groupby("Segment")["Sales"].sum()

plt.pie(
    segment_sales.values,
    labels=segment_sales.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Sales by Segment")

plt.savefig("../images/charts/sales_by_segment.png", dpi=300)

plt.show()
```

```python
subcategory_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)

subcategory_sales.head(10)
```

```python
plt.figure(figsize=(10,6))

subcategory_sales.head(10).plot(kind="bar")

plt.title("Top 10 Sub-Categories")
plt.xlabel("Sub-Category")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=30)

plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(df["Discount"], df["Profit"])

plt.title("Discount vs Profit")

plt.xlabel("Discount")

plt.ylabel("Profit")

plt.savefig("../images/charts/discount_vs_profit.png", dpi=300)

plt.show()
```
